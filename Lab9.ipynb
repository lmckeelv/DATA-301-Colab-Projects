```python
# Start your lab here

"""This dataset examines the relative abundance, or bed density, of the endangered Rocky Mountain Ridged Mussels (Gonidea angulata).
As a freshwater mussel species only found in the Okanagan Valley within Canada, seven local sites were chosen for surveying from
2011 to 2016. I intended to examine any increase in mussel density over time, view the mean mussel count surveyed in these sites,
see if any particular municipalities had more favorable mussel population growth (total and average), and then examined why.
From the data that's accessible to me, it's also possible to see the depth at which these mussels were found, which I also
analyzed in case it had any correlation with the mussel bed density. After this, I included an interactive map displaying the
average mussel count in these sites in a year of interest — 2015, which saw a massive spike in mussel density overall.
I also used BeautifulSoup to extract the data dictionary and displayed it below for easier access."""

# Name: Laura McKeel Viciana
# Student Number: 55165724

# IN-CLASS LIBRARIES:
import pandas as pd
import numpy as np # Used to view mean mussel count and std mussel count
import matplotlib.pyplot as plt
from scipy import stats # Used for linear regression
import geopandas as gpd

# NEW LIBRARIES:
import seaborn as sb
# Used seaborn to create histogram ("mussel count distribution by depth")
# and a horizontal bar chart ("avg mussel count by municipality")

import urllib.request
from bs4 import BeautifulSoup # Used to display data dictionary, and urllib.request in order to avoid c+p the html

from geopandas import GeoDataFrame # Used for interactive map
from geopandas import points_from_xy

import io # Used to prevent "passing literal html to 'read_html' is deprecated" warning

# Url of data dictionary table
url = "https://api-proxy.edh-cde.dfo-mpo.gc.ca/catalogue/records/2559aac3-fd5a-4d6d-bcf5-8cf6b9629e2f/attachments/RMRMData_Dictionary.htm"

# w/o urllib.request, had to copy/paste whole html doc (way too big)
with urllib.request.urlopen(url) as response:
    html = response.read()

soup = BeautifulSoup(html, 'html.parser')

tables = soup.find_all('table')
df_dictionary = pd.read_html(io.StringIO(str(tables[1])))[0] # From the tables, use second

print(df_dictionary.head()) # Display second table

# Url of data file
url2 = "https://api-proxy.edh-cde.dfo-mpo.gc.ca/catalogue/records/2559aac3-fd5a-4d6d-bcf5-8cf6b9629e2f/attachments/RMRM_Survey_Counts.csv"

df = pd.read_csv(url2)

df.to_csv("Mussel_Surveys.csv", index=False)

mussel_count_per_year = (
    df.groupby("Year Surveyed")["Rocky Mountain Rigid Mussel Count"]
    .sum())

mussel_count_per_year.plot(kind='bar')
plt.title("Total Mussel Count per Year")
plt.ylabel("Mussel Count")
plt.xlabel("Year")
plt.show()

df_clean = df.copy()

# Drops empty rows
df_clean.dropna(subset=["Depth Observed (m)", "Rocky Mountain Rigid Mussel Count"], inplace=True)

# Converts "Rocky Mountain Rigid Mussel Count" column to NumPy array
mussel_counts = df_clean["Rocky Mountain Rigid Mussel Count"].to_numpy()

mean_count = np.mean(mussel_counts)
std_count = np.std(mussel_counts)

print("Mean Mussel Count:", mean_count)
print("Standard Deviation", std_count)

# From the mean and std count results (standard dev. is nearly twice the mean),
# it's clear that there's been high variability in the mussel bed density throughout
#  the years. Mussel presence is extremely uneven — some sites have very low or nonexistent
# counts, and others have very high spikes ("zero-heavy data").

# After groupby, reset_index brings back regular numerical index
municipality_trends = df.groupby(["Closest Municipality", "Year Surveyed"])["Rocky Mountain Rigid Mussel Count"].sum().reset_index()

# Pivot -> new table from og dataframe
pivot_data = municipality_trends.pivot(index="Year Surveyed", columns="Closest Municipality", values="Rocky Mountain Rigid Mussel Count")

# Plot line graph for each municipality
pivot_data.plot(marker='o')
plt.title("Total Mussel Counts per Municipality Over Time")
plt.xlabel("Year Surveyed")
plt.ylabel("Total Mussel Count")
plt.legend(title="Municipality", loc='upper left')
plt.grid(True)
plt.show()


#SCIPY PART

x = df_clean["Depth Observed (m)"].to_numpy()
y = df_clean["Rocky Mountain Rigid Mussel Count"].to_numpy()

# Linear regression
slope, intercept, r_value, p_value, slope_std_error = stats.linregress(x, y)

# Predicted y-values based on regression
predicted_y = intercept + slope * x

# Prediction (residual) error
pred_error = y - predicted_y
degrees_freedom = len(x) - 2
residual_std_error = np.sqrt(np.sum(pred_error**2) / degrees_freedom)

print("Linear Regression Results:")
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)
print("P-value:", p_value)
print("Residual error:", residual_std_error)

# Seems to be a very weak relationship between depth and mussel count (R² approx. 0.059).
# While the P-value is extremely low, practically speaking, connection is weak.
# Therefore:
# From these results, data does not appear to have a strong linear relationship between depth and mussel count.
# We can conclude that mussels are most abundant at mid-range depths.

# This might mean that there an optimal depth range where conditions are most suitable for mussels,
# or that other environmental factors (temperature? nutrients? etc.) may be interacting with depth.


# SEABORN PART

# Histogram for "mussel count distribution by depth" (visually confirms conclusion from lin. regress. results)
sb.histplot(df_clean, x="Depth Observed (m)", weights="Rocky Mountain Rigid Mussel Count", bins=30, kde=True)
plt.title("Mussel Count Distribution by Depth")
plt.show()

# Avg mussel count by municipality
municipality_avg = df.groupby("Closest Municipality")["Rocky Mountain Rigid Mussel Count"].mean().reset_index()

# Sort by avg count
municipality_avg = municipality_avg.sort_values(by="Rocky Mountain Rigid Mussel Count", ascending=False)
print(municipality_avg)

# Horizontal bar chart for "avg mussel count by municipality"
sb.barplot(
    x="Rocky Mountain Rigid Mussel Count",
    y="Closest Municipality",
    data=municipality_avg,
    hue="Closest Municipality",
    palette="Set2", # Makes diff. municipalities colorful
    legend=False
)
plt.xlabel("Average Mussel Count")
plt.ylabel("Municipality")
plt.title("Average Mussel Count by Municipality")
plt.show()

# Summerland has consistently high counts, suggesting either that its location is favorable for
# the mussels, or that Summerland is the only municipality receiving consistent and thorough surveying.
# 2015 may have also marked a change in the surveying process or was affected by some environmental factor,
# as its count is especially high in general compared to other years.


#GEOPANDAS PART

df_year_filtered = df[df["Year Surveyed"] == 2015]

agg_data = df_year_filtered.groupby("Index Site").agg({
    "Rocky Mountain Rigid Mussel Count": "mean",
    "Latitude (decimal degrees)": "mean",
    "Longitude (decimal degrees)": "mean"
}).reset_index()

geometry = points_from_xy(agg_data["Longitude (decimal degrees)"], agg_data["Latitude (decimal degrees)"])
gdf = GeoDataFrame(agg_data, geometry=geometry)

canada = gpd.read_file("gadm41_CAN_1.shp")
bc = canada[canada["NAME_1"] == "British Columbia"]

# Map wasn't showing w/o proper CRS (Coordinate Reference System)
gdf.set_crs("EPSG:4326", inplace=True) #EPSG:4326 -> common CRS code

gdf["Avg. in 2015"] = ("Mussel Count: " +
                     gdf["Rocky Mountain Rigid Mussel Count"].round(2).astype(str) +
                     "<br>Site: " +
                     gdf["Index Site"].astype(str))

interactive_map = gdf.explore(
    color='purple',
    tooltip="Avg. in 2015"
)

interactive_map  # Renders interactive map
```
